```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st

plt.close("all")

df = pd.read_csv('../data/spotify_data_clean.csv')

x = pd.to_numeric(df['track_number'], errors='coerce')
y = pd.to_numeric(df['track_duration_min'], errors='coerce')

plt.scatter(x,y)
plt.xlabel("Track number")
plt.title("Track number vs track duration")
plt.ylabel("Track duration, in minutes")
plt.show()
```

```python
print('Track number: mean=%.3f stdv=%.3f' % (np.mean(x), np.std(x)))
print('Track duration: mean=%.3f stdv=%.3f' % (np.mean(y), np.std(y)))


df.head(10)
df.info()
df.describe()


# print(st.pearsonr(x,y))

```
