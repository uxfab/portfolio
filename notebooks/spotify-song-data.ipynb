```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.close("all")
```

```python
data = pd.read_csv('spotify_data_clean.csv')
df = pd.DataFrame(data)

# prints data frame
print(df.to_string())

# scatter plot duration x track number
df.plot(kind = 'scatter', x = 'track_number', y = 'track_duration_min')
plt.xlabel("Track number")
plt.ylabel("Track duration, in minutes")
plt.show()

# histogram of track numbers
df['track_number'].hist(bins=25)
plt.xlabel("Track number")
plt.ylabel("Frequency")
plt.show()
```
