```python
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt

# Reads in the 2008 lightning strike dataset
df = pd.read_csv('../data/Unicorn_Companies.csv')

# Inspects the first ten rows
df.head(10)
```

```python
df['Date Joined']= pd.to_datetime(df['Date Joined'])
```

```python
df.info()
```

```python
df['Year Joined'] = df['Date Joined'].dt.year
```

```python
df.head(10)
```

```python
df['Years to Reach'] = df['Year Joined'] - df['Year Founded']
df.head()
```

```python
df_by_industry=df.groupby('Industry')['Years to Reach'].mean().sort_values(ascending=True).head(10)
df_by_industry
```

```python
plt.bar(x=df_by_industry.index, height=df_by_industry.values, label="Years to Reach")
plt.xticks(rotation=45, ha='right')
plt.plot()

plt.xlabel("Industry")
plt.ylabel("Years (mean)")
plt.title("Mean number of years to reach unicorn status by industry")
plt.legend()
plt.show()
```
